```
import cv2 as cv
from google.colab.patches import cv2_imshow

image_directory = '/content/drive/My Drive/PUBLIC/PROJECT/ENEMY/'

# Read images
images = []
for i in range(1, 8, 1):
  file_name = 'pedestrian_0' + str(i) + '.jpg'
  file_path = image_directory + file_name
  print(file_path)
  img = cv.imread(file_path)
  cv2_imshow(img)
  images.append([img, file_name])
```

```
# HOG object detection: Human/People
hog = cv.HOGDescriptor()
hog.setSVMDetector(cv.HOGDescriptor_getDefaultPeopleDetector())

for image, name in images:
  detected, _ = hog.detectMultiScale(image)

  for (x, y, w, h) in detected:
    cv.rectangle(image, (x, y, w, h), (0, 255, 0), 3)

  print(name)
  cv2_imshow(image)
```

```
# Haar Cascade object detection: Human/People
xml_path = image_directory + 'haarcascade_fullbody.xml'
cascade = cv.CascadeClassifier(xml_path)

for image, name in images:
  detected = cascade.detectMultiScale(image)

  for (x, y, w, h) in detected:
    cv.rectangle(image, (x, y, w, h), (0, 255, 0), 3)
  
  print(name)
  cv2_imshow(image)
```
