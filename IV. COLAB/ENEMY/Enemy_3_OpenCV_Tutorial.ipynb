```
# Getting Started with Images
# https://docs.opencv.org/master/db/deb/tutorial_display_image.html

import cv2 as cv
import sys

img = cv.imread('/content/drive/My Drive/PUBLIC/PROJECT/ENEMY/starry_night.jpg')

if img is None:
  sys.exit('Could not read the image.')

cv2_imshow(img) # cv.imshow('Display window', img)
k = cv.waitKey(0)

if k == ord('s'):
  cv.imwrite('/content/drive/My Drive/PUBLIC/PROJECT/ENEMY/starry_night_write-test.jpg', img)
```

```
# Getting Started with Videos
# https://docs.opencv.org/master/dd/d43/tutorial_py_video_display.html

import numpy as np

cap = cv.VideoCapture(0)
if not cap.isOpened():
  print('Cannot open camera')
  # exit() # commented out due to colab run-time downed

while True:
  ret, frame = cap.read()

  if not ret:
    print("Can't receive frame (stread end?). Exiting ...")
    break
  
  gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
  cv.imshow('frame', gray)
  if cv.waitKey(1) == ord('q'):
    break

cap.release()
cv.destroyAllWindows()

# How to capture video on Colab (Web Browser)
# https://stackoverflow.com/questions/54389727/how-to-use-cap-cv2-videocapture0-in-google-colab
# https://colab.research.google.com/notebooks/snippets/advanced_outputs.ipynb#scrollTo=9OyC1_bSEccg
```

```
cap = cv.VideoCapture(0)

fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('output.avi', fourcc, 20.0, (640, 480))

while cap.isOpened():
  ret, frame = cap.read()
  if not ret:
    print("Can't receive frame (stream end?). Exiting ...")
    break
  frame = cv.flip(frame, 0)

  out.write(frame)

  cv2_imshow(frame)
  if cv.waitKey(1) == ord('q'):
    break

cap.release()
out.release()
cv.destroyAllWindows()
```
